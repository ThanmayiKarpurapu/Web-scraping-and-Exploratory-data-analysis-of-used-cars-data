```python
#IMPORTING LIBRARIES
import warnings
warnings.filterwarnings('ignore')
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import requests
from bs4 import BeautifulSoup
import re
import time
```

```python
#FINAL WEB SCRAPINGCODE
Title=[]
Price=[]
Brand=[]
Fuel_type=[]
Kilometers=[]
City=[]
Year=[]
Model = []
#https://www.cartrade.com/buy-used-cars/delhi/c/#city=10&sc=-1&so=-1&pn=1
for i in range(1,20):
    url= f'https://www.cartrade.com/buy-used-cars/delhi/c/#city=10&sc=-1&so=-1&pn={i}'
    response=requests.get(url)
    soup=BeautifulSoup(response.text)
    
    container=soup.find_all("div",class_="details_out")
    for i in container:
        title=i.find("h2",class_="h2heading truncate")
        
        regex=re.findall("\d+\s((?:Toyota|Tata|Datsun|Maruti Suzuki|Mercedes-Benz|BMW|Ford|Honda|Lexus|Bentley|Volvo|Audi|Hyundai|Nissan|Skoda|Mahindra|Porsche|Jaguar|Kia|Land Rover|MG|MINI Cooper|Land Rover|Volkswagen|Jeep|Volks Wagon|Renault))",title.text)
        if regex:
            Brand.append(regex[0])
        else:
            Brand.append(np.nan)
    for i in container:
        model = i.find("h2",class_= "h2heading truncate")
        regex= re.findall("\d+\s(?:Toyota|Tata|Datsun|Maruti Suzuki|Mercedes-Benz|BMW|Ford|Lexus|Honda|Audi|Hyundai|Bentley|Volvo|Nissan|Skoda|Mahindra|Porsche|Jaguar|Kia|Land Rover|MG|MINI Cooper|Land Rover|Volkswagen|Jeep|Volks Wagon|Renault)(.*)\s",model.text)
        if regex:
            Model.append(regex[0])
                
    for i in container:
        price=i.find("div",class_="cr_prc")
        text=price.text
        #print(text)
        regex=re.findall("(\d+,?\d+,?\d+)",text)
        if regex:
            Price.append(regex[0])
        else:
            Price.append(np.nan)
            
    for i in container:
        kilometers= i.find("div",class_="info_cr_new")
        text1=kilometers.text
        #print(text1)
        regex=re.findall("(\d+?,\d+)\sKMs",text1)
        if regex:
            Kilometers.append(regex[0])
        else:
            Kilometers.append(np.nan)
            
            
    for i in container:
        fuel_type= i.find("div",class_="info_cr_new")
        text2=fuel_type.text
        regex=re.findall("KMs\W(\w+)",text2)
        if regex:
            Fuel_type.append(regex[0])
        else:
            Fuel_type.append(np.nan)
            
    for i in container:
        city=i.find("div",class_="info_cr_new")
        text3=city.text
        regex=re.findall("(\w+)$",text3)
        if regex:
            City.append(regex[0])
        else:
            City.append(np.nan)
            
    for i in container:
        year= i.find("h2",class_="h2heading truncate")
        print(year.text)
        regex=re.findall("\W+(\d+).*",year.text)
        if regex:
            Year.append(regex[0])
        else:
            Year.append(np.nan)
                
        
     
```
